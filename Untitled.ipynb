```python
# Assignment 1: Implementing Feedforward Neural Network in Keras + TensorFlow
# Dataset: Fashion MNIST

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# 2.1 Load the Fashion MNIST dataset
# Each image is 28x28 grayscale, 10 clothing categories
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# 2.2 Preprocess / Normalize the data
# Convert pixel values from [0,255] to [0,1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# Flatten 28x28 images into 784-dimensional vectors for FFN
x_train = x_train.reshape(-1, 28 * 28)
x_test  = x_test.reshape(-1, 28 * 28)

# 2.3 Build the Feedforward Neural Network model
model = keras.Sequential([
    layers.Input(shape=(784,)),           # Input layer: 784 features
    layers.Dense(256, activation="relu"), # Hidden layer 1
    layers.Dense(128, activation="relu"), # Hidden layer 2
    layers.Dense(10, activation="softmax")# Output layer: 10 classes
])

# Show model architecture
model.summary()

# 2.4 Compile the model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",  # because labels are integers 0–9
    metrics=["accuracy"]
)

# 2.5 Train the model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1,  # 10% of training data used as validation
    verbose=1
)

# 2.6 Evaluate the model on test data
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nTest Accuracy: {test_acc:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# 2.7 Make predictions on first 5 test images
import numpy as np

pred_probs = model.predict(x_test[:5])    # prediction probabilities
pred_labels = np.argmax(pred_probs, axis=1)

print("\nPredicted labels for first 5 test samples:", pred_labels)
print("True labels:                          ", y_test[:5])
```

```python
# import cv2
# import os
# import numpy as np

# # ---- FIX THIS PATH ----
# DATASET_DIR = r"Face assign2/Assignmentdl2"

# print("Dataset Found:", os.path.exists(DATASET_DIR))

# person_names = ["modi", "trump"]
# label_map = {0: "Modi", 1: "Trump"}

# faces = []
# labels = []

# face_cascade = cv2.CascadeClassifier(
#     cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
# )

# # -------- Load Images --------
# for label, person in enumerate(person_names):
#     folder_path = os.path.join(DATASET_DIR, person)

#     print("Checking:", folder_path, "Exists:", os.path.exists(folder_path))

#     for filename in os.listdir(folder_path):
#         img_path = os.path.join(folder_path, filename)
#         img = cv2.imread(img_path)

#         if img is None:
#             print("Skipped (can't read):", img_path)
#             continue

#         gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#         faces_detected = face_cascade.detectMultiScale(gray, 1.1, 5)

#         # If no face detected from image, skip silently
#         for (x, y, w, h) in faces_detected:
#             faces.append(gray[y:y+h, x:x+w])
#             labels.append(label)
#             break

# print(f"Faces Loaded for Training: {len(faces)}")

# # -------- Train recognizer --------
# recognizer = cv2.face.LBPHFaceRecognizer_create()
# recognizer.train(faces, np.array(labels))
# print("Training Finished.")

# # -------- Webcam Recognition --------
# cap = cv2.VideoCapture(0)

# while True:
#     ret, frame = cap.read()
#     if not ret:
#         break

#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#     detected = face_cascade.detectMultiScale(gray, 1.1, 5)

#     for (x, y, w, h) in detected:
#         face_crop = gray[y:y+h, x:x+w]
#         label, confidence = recognizer.predict(face_crop)

#         name = label_map[label]
#         text = f"{name} ({int(confidence)})"

#         cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
#         cv2.putText(frame, text, (x, y - 10),
#                     cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

#     cv2.imshow("Narendra Modi vs Donald Trump Recognition", frame)

#     if cv2.waitKey(10) & 0xFF == ord('a'):
#         break

# cap.release()
# cv2.destroyAllWindows()




# Assignment 2: Facial Recognition using OpenCV (No Webcam)
# Binary classification: Narendra Modi vs Donald Trump

import cv2
import os
import numpy as np

# -----------------------------
# 1. Setup paths and labels
# -----------------------------
# Change this to your dataset folder path
# Example: r"C:\Users\yash\Desktop\DL practicals\Face assign2\Assignmentdl2"
DATASET_DIR = r"Assignmentdl2"

person_names = ["modi", "trump"]   # folder names
label_map = {0: "Modi", 1: "Trump"}

# -----------------------------
# 2. Load images and prepare data
# -----------------------------
faces = []   # list of face images (grayscale)
labels = []  # list of integer labels (0 or 1)

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

for label, person in enumerate(person_names):
    person_folder = os.path.join(DATASET_DIR, person)

    if not os.path.exists(person_folder):
        print("Folder not found:", person_folder)
        continue

    for img_name in os.listdir(person_folder):
        img_path = os.path.join(person_folder, img_name)

        # read image
        img = cv2.imread(img_path)
        if img is None:
            print("Cannot read:", img_path)
            continue  # skip broken files

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # detect face in the image (we assume 1 main face per image)
        detected = face_cascade.detectMultiScale(gray, 1.1, 5)
        if len(detected) == 0:
            print("No face found in:", img_path)
            continue

        for (x, y, w, h) in detected:
            face_roi = gray[y:y+h, x:x+w]
            # resize to fixed size to be safe
            face_roi = cv2.resize(face_roi, (200, 200))
            faces.append(face_roi)
            labels.append(label)
            break  # take only first detected face

print(f"Total faces loaded: {len(faces)}")

faces = np.array(faces)
labels = np.array(labels)

# -----------------------------
# 3. Split into train & test
#    (here very simple: first half train, second half test)
#    If you add more images later, this still works.
# -----------------------------
if len(faces) < 2:
    print("Warning: Very few images, using all for both train and test.")
    faces_train, faces_test = faces, faces
    labels_train, labels_test = labels, labels
else:
    split_idx = len(faces) // 2  # simple 50-50 split
    faces_train, faces_test = faces[:split_idx], faces[split_idx:]
    labels_train, labels_test = labels[:split_idx], labels[split_idx:]

print("Train samples:", len(faces_train))
print("Test samples :", len(faces_test))

# -----------------------------
# 4. Create and train recognizer
# -----------------------------
# Requires: pip install opencv-contrib-python
recognizer = cv2.face.LBPHFaceRecognizer_create()

recognizer.train(list(faces_train), np.array(labels_train))
print("Training completed.")

# Optional: save model
recognizer.write("modi_trump_lbph_model.yml")

# -----------------------------
# 5. Test on test set
# -----------------------------
correct = 0

for i, test_face in enumerate(faces_test):
    label_true = labels_test[i]
    # LBPH expects grayscale image
    label_pred, confidence = recognizer.predict(test_face)

    name_true = label_map.get(label_true, "Unknown")
    name_pred = label_map.get(label_pred, "Unknown")

    print(f"Sample {i+1}: True = {name_true}, Pred = {name_pred}, Conf = {int(confidence)}")

    if label_true == label_pred:
        correct += 1

accuracy = correct / len(faces_test) if len(faces_test) > 0 else 0
print(f"\nTest Accuracy: {accuracy * 100:.2f}%")

# -----------------------------
# 6. (Optional) Show test images with prediction
# -----------------------------
for i, test_face in enumerate(faces_test):
    label_pred, confidence = recognizer.predict(test_face)
    name_pred = label_map.get(label_pred, "Unknown")

    show_img = cv2.resize(test_face, (300, 300))
    cv2.putText(show_img, f"{name_pred}", (10, 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

    cv2.imshow(f"Test Sample {i+1}", show_img)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```

# Assignment 3

```python
# Assignment 3: Image Classification using CNNs (Multiclass)
# Dataset: CIFAR-10 (10 classes, 32x32 color images)

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt

# 1. Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

print("Train images:", x_train.shape)
print("Train labels:", y_train.shape)
print("Test images:", x_test.shape)
print("Test labels:", y_test.shape)

# Save raw labels for later display (since we’ll one-hot encode)
y_train_raw = y_train.copy()
y_test_raw = y_test.copy()

# Class names for CIFAR-10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# 2. Preprocessing
# 2.1 Normalize pixel values to [0,1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# 2.2 One-hot encode labels (10 classes)
y_train = to_categorical(y_train, num_classes=10)
y_test  = to_categorical(y_test, num_classes=10)

# 3. Build CNN model
model = models.Sequential()

# Convolution + Pooling block 1
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))

# Convolution + Pooling block 2
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Convolution block 3
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Flatten + Dense layers for classification
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dropout(0.5))  # to reduce overfitting
model.add(layers.Dense(10, activation='softmax'))  # 10 classes

# Show summary
model.summary()

# 4. Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  # multiclass
    metrics=['accuracy']
)

# 5. Train the model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=64,
    validation_data=(x_test, y_test),
    verbose=1
)

# 6. Evaluate on test set
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nTest accuracy: {test_acc:.4f}")
print(f"Test loss: {test_loss:.4f}")

# 7. Plot training & validation accuracy and loss
plt.figure(figsize=(10, 4))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 8. Predict on some test images
num_samples = 5
sample_images = x_test[:num_samples]
sample_labels = y_test_raw[:num_samples].flatten()

pred_probs = model.predict(sample_images)
pred_classes = np.argmax(pred_probs, axis=1)

# Show images with predicted & true labels
plt.figure(figsize=(10, 3))
for i in range(num_samples):
    plt.subplot(1, num_samples, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(sample_images[i])
    plt.xlabel(f"P: {class_names[pred_classes[i]]}\nT: {class_names[sample_labels[i]]}")
plt.tight_layout()
plt.show()
```

```python
# Assignment 4 (Improved): Time Series Weather Forecasting using LSTM (RNN)
# Goal: Predict RainTomorrow (Yes/No) using multiple weather features

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# -----------------------------------------
# 1. Load Dataset
# -----------------------------------------
file_path = "rain_forecasting assign4.csv"   # update if needed
data = pd.read_csv(file_path)

print("Columns:", data.columns)

# -----------------------------------------
# 2. Select Features & Target
# -----------------------------------------
feature_cols = [
    'MinTemp', 'MaxTemp',
    'Humidity9am', 'Humidity3pm',
    'Pressure9am', 'Pressure3pm',
    'WindSpeed9am', 'WindSpeed3pm',
    'RainToday'
]

target_col = 'RainTomorrow'

# Keep only needed columns
data = data[feature_cols + [target_col]]

# Drop rows with missing values for simplicity
data = data.dropna().reset_index(drop=True)

# Encode RainToday & RainTomorrow (Yes/No → 1/0)
le = LabelEncoder()
data['RainToday'] = le.fit_transform(data['RainToday'])       # No=0, Yes=1
data['RainTomorrow'] = le.fit_transform(data['RainTomorrow']) # No=0, Yes=1

# Features and target arrays
X_all = data[feature_cols].values
y_all = data[target_col].values  # shape: (N,)

# -----------------------------------------
# 3. Scale Features
# -----------------------------------------
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_all)

# -----------------------------------------
# 4. Create Time Sequences for LSTM
#    Use past `time_steps` days to predict RainTomorrow of next day
# -----------------------------------------
def create_sequences(features, labels, time_steps=7):
    X_seq, y_seq = [], []
    for i in range(time_steps, len(features)):
        X_seq.append(features[i-time_steps:i])  # window of past days
        y_seq.append(labels[i])                 # label of current day
    return np.array(X_seq), np.array(y_seq)

time_steps = 7  # last 7 days
X, y = create_sequences(X_scaled, y_all, time_steps)

print("X shape:", X.shape)  # (samples, time_steps, features)
print("y shape:", y.shape)

# Train-test split (80% / 20%)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# -----------------------------------------
# 5. Build Improved LSTM Model
# -----------------------------------------
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(time_steps, X.shape[2])),
    Dropout(0.3),
    LSTM(32),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # binary classification (rain or not)
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# -----------------------------------------
# 6. Train Model
# -----------------------------------------
history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

# -----------------------------------------
# 7. Evaluate & Metrics
# -----------------------------------------
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

acc = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {acc:.4f}\n")

print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=['NoRain', 'Rain']))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Assignment 6: Sentiment Analysis using LSTM
# Dataset: IMDB Movie Reviews (Keras)
# Classes: 0 = Negative, 1 = Positive

import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# 1. Load IMDB dataset (top 10,000 words)
max_features = 10000      # vocabulary size
max_len = 500             # pad / truncate reviews to 500 words

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape :", X_test.shape, y_test.shape)

# 2. Pad sequences so all reviews have same length
X_train = sequence.pad_sequences(X_train, maxlen=max_len)
X_test  = sequence.pad_sequences(X_test,  maxlen=max_len)

print("Padded train shape:", X_train.shape)
print("Padded test shape :", X_test.shape)

# 3. Build LSTM model
model = Sequential()
model.add(Embedding(max_features, 128, input_length=max_len))  # word embeddings
model.add(LSTM(128))                                           # LSTM layer
model.add(Dense(1, activation="sigmoid"))                      # binary output

model.summary()

# 4. Compile model
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# Early stopping (optional but good)
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True
)

# 5. Train model
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# 6. Evaluate on test data
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy: {test_acc:.4f}")
print(f"Test Loss    : {test_loss:.4f}")

# 7. Predict sentiment for first 5 test reviews
y_pred_prob = model.predict(X_test[:5])
y_pred = (y_pred_prob > 0.5).astype("int32").flatten()

print("\nFirst 5 predictions (0=Negative, 1=Positive)")
print("Predicted:", y_pred)
print("Actual   :", y_test[:5])
```
